```julia
using Flux
using Flux, Statistics
using Flux: onehotbatch, onecold, @epochs
using Flux.Losses: logitcrossentropy
using Base: @kwdef
```

```julia
NUM_CLASSES = 10;
IMAGE_SIZE = (28,28,1);
DEVICE = cpu;
PRE_PROCESS = true;
EPOCHS=1;
```

```julia
Q_LAYERS = 1
CONV_SIZE = 4
GATES_PER_LAYER = 6
qpu = devices.LocalSimulator();
```

```julia
function random_layer(target_circuit)
    single_qubit_gates = [circuit.Circuit().rx, circuit.Circuit().ry, circuit.Circuit().rz]
    two_qubit_gates = [circuit.Circuit().cnot]
    
    for layer in 1:Q_LAYERS
        for i in 1:GATES_PER_LAYER
            prob = rand()
            gate = nothing
            if prob > 0.3
                random_gate = rand(single_qubit_gates)
                control_qubit = rand(0:(CONV_SIZE-1))
                angle = pi * rand()
                gate = random_gate(control_qubit, angle)
            else
                random_gate = rand(two_qubit_gates)
                control_qubit = rand(0:(CONV_SIZE-1))
                target_qubit = rand(0:(CONV_SIZE-1))
                while control_qubit == target_qubit
                    target_qubit = rand(0:(CONV_SIZE-1))
                end 
                gate = random_gate(control_qubit, target_qubit)
            end
            target_circuit.add(gate)
        end
    end;
    
    return target_circuit
end;
```

```julia
rand_params = [0.0444, 0.145, 3.3, 2.51] #rand(-pi:pi, 4)
function random_layer(target_circuit)
    target_circuit.add(circuit.Circuit().rx(0, rand_params[1]))
    target_circuit.add(circuit.Circuit().rz(3, rand_params[2]))
    target_circuit.add(circuit.Circuit().cnot(2,1))
    target_circuit.add(circuit.Circuit().cnot(3,0))
    target_circuit.add(circuit.Circuit().ry(0,rand_params[3]))
    target_circuit.add(circuit.Circuit().rx(0,rand_params[4]))
    
    return target_circuit
end;
```

```julia
function get_filter(phi)
    conv_filter = circuit.Circuit()
    
    for wire in 1:CONV_SIZE
        conv_filter.add(circuit.Circuit().ry(wire - 1, pi * phi[wire]))
    end

    conv_filter = random_layer(conv_filter)
    
    conv_filter.expectation(observable=circuit.Observable.Z())
    #conv_filter.expectation(observable=circuit.Observable.Z(), target=[1])
    #conv_filter.expectation(observable=circuit.Observable.Z(), target=[2])
    #conv_filter.expectation(observable=circuit.Observable.Z(), target=[3])
    
    return conv_filter 
end;
```

```julia
println(get_filter([0.05490196,0.00392157,0.60392157,0.99215686]).diagram())
qpu.run(get_filter([0.05490196,0.00392157,0.60392157,0.99215686]), shots=0).result().result_types[1].value
```

```julia
function quanv(image)
    out = zeros(14, 14, 4)
    for j = range(1, 28, step=2)
        conv_filter = nothing
        for k = range(1, 28, step=2)
            #@show j, k
            conv_filter = get_filter([
                    image[j, k],
                    image[j, k + 1],
                    image[j + 1, k],
                    image[j + 1, k + 1]
            ])
            #@show conv_filter
            qonv = qpu.run(conv_filter, shots=0).result().result_types[1].value
            #@show qonv
            for c in 1:4
                #@show c, qonv[c], [image[j, k],image[j, k + 1],image[j + 1, k],image[j + 1, k + 1]]
                out[ceil(Int32, j/2), ceil(Int32, k/2), c] = qonv[c]
            end
        end
    end
    
    return out
end
```

```julia
qxtrain = zeros(14, 14, 4, TRAIN_SIZE)
qxtest = zeros(14, 14, 4, TEST_SIZE)

plots = []

for i in 1:TRAIN_SIZE
    qxtrain[:,:,:, i] = quanv(xtrain[:,:,i])
end

for i in 1:TEST_SIZE
    qxtest[:,:,:, i] = quanv(xtest[:,:,i])
end
```
